```python
# Inspecci√≥n de tr√°fico
print("Shape:", trafico_tabla.shape)
print("\nTipos de columnas:")
print(trafico_tabla.dtypes)
print("\nValores nulos:")
print(trafico_tabla.isnull().sum())
print("\nPrimeras filas:")
print(trafico_tabla.head())
```

```python
import datetime

# A) Eliminar fila "Total"
trafico_tabla_clean = trafico_tabla[trafico_tabla['Traffic source'] != "Total"]

# B) Rellenar nulos con 0 en Impressions y CTR
trafico_tabla_clean['Impressions'] = trafico_tabla_clean['Impressions'].fillna(0)
trafico_tabla_clean['Impressions click-through rate (%)'] = trafico_tabla_clean['Impressions click-through rate (%)'].fillna(0)

# C) Convertir 'Average view duration' de HH:MM:SS a minutos
def duration_to_minutes(duration_str):
    x = datetime.datetime.strptime(duration_str, '%H:%M:%S')
    return x.minute + x.second/60 + x.hour * 60

trafico_tabla_clean['Average view duration (min)'] = trafico_tabla_clean['Average view duration'].apply(duration_to_minutes)

# D) Verificar estado final
print("Shape final:", trafico_tabla_clean.shape)
print("\nValores nulos finales:")
print(trafico_tabla_clean.isnull().sum())
print("\nPrimeras filas limpias:")
print(trafico_tabla_clean.head())

```

1- Eliminamos la fila "Total"
Es una fila de resumen global. Si la dejamos, distorsiona c√°lculos como promedios o porcentajes por fuente.
Queremos analizar fuente por fuente, no mezclar con el total.

2- Tratamos los valores nulos (NaN)
Ejemplo: Shorts feed no genera impresiones como los videos normales, por eso aparecen NaN.
Si dejamos los NaN, no podemos sumar o promediar sin errores. Rellenamos con 0 porque significa ‚Äúno aplica‚Äù.

3- Convertimos Average view duration (0:08:11) a minutos decimales
Con el formato texto 0:08:11 es dif√≠cil hacer c√°lculos (promedios, comparaciones).
Al convertirlo a minutos (por ejemplo, 0:08:11 ‚Üí 8.18 min), podemos analizar cu√°l fuente de tr√°fico tiene mejor retenci√≥n.

```python
import numpy as np

# 1) Partimos de resumen_tabla_clean
df = resumen_tabla_clean.copy()

# 2) Asegurar que no hay ceros o nulos en las variables clave
df = df[(df['Views'] > 0) & (df['Duration'] > 0)]

# 3) Calcular duraci√≥n del video en horas y el tiempo promedio visto por vista
df['duration_h'] = df['Duration'] / 60.0
df['avg_view_duration_h'] = df['Watch time (hours)'] / df['Views']

# 4) Retenci√≥n estimada: (tiempo promedio visto) / (duraci√≥n del video)
df['retention_pct_est'] = (df['avg_view_duration_h'] / df['duration_h']) * 100

# 5) (Opcional) Limpiar outliers imposibles (>100%) si quieres
df['retention_pct_est'] = df['retention_pct_est'].clip(upper=100)

# 6) Ordenar y tomar Top 10
top10_retention = (
    df.sort_values('retention_pct_est', ascending=False)
      .loc[:, ['Video title', 'Views', 'Duration', 'Watch time (hours)', 'retention_pct_est']]
      .head(10)
      .reset_index(drop=True)
)

# (Opcional) Bottom 10 para detectar problemas
bottom10_retention = (
    df.sort_values('retention_pct_est', ascending=True)
      .loc[:, ['Video title', 'Views', 'Duration', 'Watch time (hours)', 'retention_pct_est']]
      .head(10)
      .reset_index(drop=True)
)

# 7) Mostrar resultados
print("TOP 10 por retenci√≥n estimada (%)")
display(top10_retention.style.format({'Duration':'{:.0f} min', 
                                      'Watch time (hours)':'{:.2f}',
                                      'retention_pct_est':'{:.1f}%'}))

print("\nBOTTOM 10 por retenci√≥n estimada (%)")
display(bottom10_retention.style.format({'Duration':'{:.0f} min', 
                                         'Watch time (hours)':'{:.2f}',
                                         'retention_pct_est':'{:.1f}%'}))
```

```python
import pandas as pd
import numpy as np

# --- Partimos de resumen_tabla_clean ---
df = resumen_tabla_clean.copy()

# (Re)calcular m√©tricas necesarias
df = df[(df['Views'] > 0) & (df['Duration'] > 0)]
df['avg_view_duration_h']   = df['Watch time (hours)'] / df['Views']
df['avg_view_duration_min'] = df['avg_view_duration_h'] * 60
df['avg_view_duration_sec'] = df['avg_view_duration_h'] * 3600

# Retenci√≥n estimada (%)
df['retention_pct_est'] = (df['avg_view_duration_sec'] / df['Duration']) * 100
df['retention_pct_est'] = df['retention_pct_est'].clip(upper=100)

# Clasificaci√≥n por duraci√≥n (Duration est√° en **segundos**)
def clasificar_video(d):
    if d < 300:          # < 5 min
        return 'Corto (<5 min)'
    elif d <= 2400:      # 5 a 40 min
        return 'Mediano (5-40 min)'
    else:                # > 40 min
        return 'Largo (>40 min)'

df['categoria'] = df['Duration'].apply(clasificar_video)

# Agregados por categor√≠a
resumen_categorias = (
    df.groupby('categoria')
      .agg(
          n_videos                       = ('Content', 'count'),
          total_views                    = ('Views', 'sum'),
          total_watch_time_hours         = ('Watch time (hours)', 'sum'),
          avg_duration_min               = ('Duration', lambda x: (x.mean()/60)),
          mean_retention_pct             = ('retention_pct_est', 'mean'),
          median_retention_pct           = ('retention_pct_est', 'median'),
          views_por_video_prom           = ('Views', 'mean'),
          watch_time_horas_por_video_prom= ('Watch time (hours)', 'mean')
      )
      .reset_index()
      .sort_values('total_watch_time_hours', ascending=False)
)

# Mostrar tabla formateada
resumen_categorias_style = resumen_categorias.style.format({
    'total_views': '{:,.0f}',
    'total_watch_time_hours': '{:,.1f}',
    'avg_duration_min': '{:,.1f}',
    'mean_retention_pct': '{:,.1f}%',
    'median_retention_pct': '{:,.1f}%',
    'views_por_video_prom': '{:,.0f}',
    'watch_time_horas_por_video_prom': '{:,.2f}'
})

resumen_categorias_style
```

# C√≥digo para la Retenci√≥n Media Ponderada

```python
# --- Retenci√≥n ponderada por vistas ---
df = resumen_tabla_clean.copy()

# Tiempo promedio por espectador (en horas)
df['avg_view_duration_h'] = df['Watch time (hours)'] / df['Views']
df['avg_view_duration_sec'] = df['avg_view_duration_h'] * 3600

# Retenci√≥n estimada
df['retention_pct_est'] = (df['avg_view_duration_sec'] / df['Duration']) * 100
df['retention_pct_est'] = df['retention_pct_est'].clip(upper=100)

# Clasificar por categor√≠a de duraci√≥n
def clasificar_video(d):
    if d < 300: return 'Corto (<5 min)'
    elif d <= 2400: return 'Mediano (5-40 min)'
    else: return 'Largo (>40 min)'

df['categoria'] = df['Duration'].apply(clasificar_video)

# Retenci√≥n ponderada
retencion_ponderada = (
    df.groupby('categoria')
      .apply(lambda x: np.average(x['retention_pct_est'], weights=x['Views']))
      .reset_index(name='retencion_ponderada_%')
)

print("Retenci√≥n media ponderada por categor√≠a:")
retencion_ponderada
```

```python
# Clasificaci√≥n b√°sica de programa seg√∫n t√≠tulo
def clasificar_programa(titulo):
    t = titulo.lower()
    if "bhagavad" in t:
        return "Bhagavad Gita"
    elif "domingo" in t or "bhakti" in t:
        return "Domingos de Bhakti"
    elif "veda" in t or "vedas" in t:
        return "Los Vedas"
    else:
        return "Otros"

df['programa'] = df['Video title'].apply(clasificar_programa)

# Resumen por programa
resumen_programas = (
    df.groupby('programa')
      .agg(
          n_videos=('Content','count'),
          total_views=('Views','sum'),
          total_watch_hours=('Watch time (hours)','sum'),
          avg_duration_min=('Duration', lambda x: (x.mean()/60)),
          mean_retention_pct=('retention_pct_est','mean')
      )
      .reset_index()
      .sort_values('total_watch_hours', ascending=False)
)

print("KPIs por programa:")
resumen_programas
```
